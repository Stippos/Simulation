```python
import numpy as np
import matplotlib.pyplot as plt

capacity = 164
iterations = 1000

mu = 180
sigma = 30

lowFare = 70
highFare = 100

expectedRevenues = []
deviations = []

for R in range(0, capacity + 1):
    revenues = []
    demands = np.random.normal(mu, sigma, iterations)
    
    for i in demands:
        if i <= R:
            revenues.append(lowFare * i)
        else if i <= capacity:
            revenues.append(lowFare * R + 0.6 * (i - R) * highFare)
            
    expectedRevenues.append(np.average(revenues))
    deviations.append(np.std(revenues))
    
plt.plot(expectedRevenues)
plt.plot(deviations)            
        
    
```
